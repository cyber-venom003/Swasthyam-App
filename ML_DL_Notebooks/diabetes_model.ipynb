```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import xgboost
from xgboost import XGBClassifier
from mlxtend.preprocessing import minmax_scaling 
```

```
diabetes=pd.read_csv('diabetes.csv')
diabetes
```

```
data=pd.read_csv('diabetes.csv')
del data['Outcome']
X=data
Y=diabetes['Outcome']
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```
list1=np.arange(10,1000,10)
scores=[]
for k in list1:
  model=RandomForestClassifier(n_estimators=k)
  model.fit(x_train,y_train)
  scores.append(model.score(x_test,y_test))

optimal_k=list1[scores.index(max(scores))]
print(optimal_k)
```

```
best_model=RandomForestClassifier(n_estimators=optimal_k)
best_model.fit(x_train,y_train)
best_model.score(x_test,y_test)
```

```
joblib.dump(best_model,'diabetes_model.pkl')
```

```
estimator = XGBClassifier(max_depth=1)
params={
    'n_estimators': range(50, 400, 40),
    'learning_rate': [0.1, 0.5, 0.01, 0.05]
}

grid_search = GridSearchCV(
    estimator=estimator,
    param_grid=params,
    scoring = 'roc_auc',
    cv = 10,
    verbose=True
)

grid_search.fit(x_train,y_train)
```

```
print(grid_search.best_estimator_)
```

```
best_model=grid_search.best_estimator_
best_model.fit(x_train,y_train)
best_model.score(x_test,y_test)
```

```
scaled_data = minmax_scaling(diabetes,columns=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'])
```

```
import tensorflow as tf
from keras.layers import Input, Lambda,LSTM, Dense, Flatten,Conv2D,MaxPooling2D,Dropout,Flatten,BatchNormalization
from keras.models import Model,Sequential
```

```
input_dim = x_train.shape[1]
model = Sequential()
model.add(Input(shape=(input_dim,)))
model.add(Dense(8,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
print(model.summary())

optimizer = tf.keras.optimizers.RMSprop(0.01)
```

```
model.compile(optimizer=optimizer,loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(scaled_data,diabetes['Outcome'],epochs=1000,verbose=2,validation_split=0.2)
```
