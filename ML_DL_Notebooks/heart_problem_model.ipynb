```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```
heart=pd.read_csv('heart.csv')
heart
```

```
print(len(heart[heart['target']==1]))
print(len(heart[heart['target']==0]))
```

```
data=pd.read_csv('heart.csv')
```

```
del data['target']
data
```

```
X=data
Y=heart['target']
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```
list1=np.arange(10,1000,10)
scores=[]
for k in list1:
  model=RandomForestClassifier(n_estimators=k)
  model.fit(x_train,y_train)
  scores.append(model.score(x_test,y_test))

optimal_k=list1[scores.index(max(scores))]
print(optimal_k)
```

```
clf=RandomForestClassifier(n_estimators=10)
clf.fit(x_train,y_train)
print(clf.score(x_test,y_test))
```

```
model_best=RandomForestClassifier(n_estimators=10)
model_best.fit(X,Y)
```

```
import pickle
from sklearn.externals import joblib

joblib.dump(model_best,'heart.pkl')
```
